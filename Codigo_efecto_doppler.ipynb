```python
import matplotlib.pyplot as plt
import pandas as pd
from io import StringIO

def main():
    # Page the data here
    datos = """
Frecuencia (Hz)	Nivel (dB)
43.066406	-76.158424
86.132812	-57.798279
129.199219	-45.157887
172.265625	-41.002880
215.332031	-42.494282
258.398438	-44.843796
301.464844	-44.789185
344.531250	-44.589146
387.597656	-49.495464
430.664062	-50.489651
"""
    df = pd.read_csv(StringIO(datos), sep="\t")

    plt.figure(figsize=(12, 6))
    plt.plot(df['Frecuencia (Hz)'], df['Nivel (dB)'], marker='o', linestyle='-', color='mediumblue')
    plt.title('Efecto Doppler: Sound Level vs Frecuency')
    plt.xlabel('Frecuency (Hz)')
    plt.ylabel('Level (dB)')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```
