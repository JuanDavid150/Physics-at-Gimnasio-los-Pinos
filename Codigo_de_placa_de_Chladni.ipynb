```python
import numpy as np
import matplotlib.pyplot as plt

def chladni_pattern(frequency, size=500):
    x = np.linspace(-1, 1, size)
    y = np.linspace(-1, 1, size)
    X, Y = np.meshgrid(x, y)

    n = int(frequency // 70) + 1
    m = int(frequency // 55) + 1

    Z = (np.sin(n * np.pi * X) * np.sin(m * np.pi * Y) +
         np.sin(m * np.pi * X) * np.sin(n * np.pi * Y) +
         0.5 * np.sin((n + m) * np.pi * X * Y))

    plt.figure(figsize=(6, 6))
    plt.contour(X, Y, Z, levels=[0], colors='white', linewidths=1.2)
    plt.imshow(Z, cmap='binary', extent=[-1, 1, -1, 1], alpha=0.5)
    plt.title(f"Simulated Chladni pattern:\nFrecuency: {frequency:.1f} Hz")
    plt.axis('off')
    plt.show()

if __name__ == "__main__":
    try:
        freq = float(input("Enter the frequency (in Hz): "))
        chladni_pattern(freq)
    except ValueError:
        print("Please enter a valid number")
```
