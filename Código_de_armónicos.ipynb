```python
import numpy as np
import matplotlib.pyplot as plt

def calcular_armonicos(fundamental, max_armonico, duracion=1, muestreo=1000):
    t = np.linspace(0, duracion, int(duracion * muestreo))
    señal = np.sin(2 * np.pi * fundamental * t)

    for n in range(2, max_armonico + 1):
        señal += (1/n) * np.sin(2 * np.pi * fundamental * n * t)
    return t, señal

def main():
    frecuencia_fundamental = float(input("Enter the fundamental frequency in Hz: "))
    max_armonico = int(input("Enter the maximum number of harmonics you wish to display: "))

    t, señal = calcular_armonicos(frecuencia_fundamental, max_armonico)

    plt.figure(figsize=(10, 6))
    plt.plot(t, señal)
    plt.title(f"Harmonic series of the fundamental frequency {frecuencia_fundamental} Hz")
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitud')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
